```python
# import dependencies
import pandas as pd
from pandas.plotting import scatter_matrix

import matplotlib as mlp
from matplotlib import pyplot as plt

from sklearn.impute import SimpleImputer
```

```python
life_expectancy_df = pd.read_csv("data/Life Expectancy Data.csv")
life_expectancy_df
```

```python
# inspect data types and check for null-values using pandas info method
life_expectancy_df.info()
```

```python
# examine column names (look for extra spacing)
life_expectancy_df.columns
```

```python
# remove extra space from "Life expectancy " using pandas rename
life_expectancy_df.rename(columns={
    "Life expectancy ": "Life expectancy",
    "Measles ": "Measles",
    " BMI ": "BMI",
    "under-five deaths ": "under-five deaths",
    "Diphtheria ": "Diphtheria",
    " HIV/AIDS": "HIV/AIDS",
    " thinness  1-19 years": "thinness 1-19 years",
    " thinness 5-9 years": "thinness 5-9 years"
}, inplace=True)

life_expectancy_df.columns
```

```python
# df['DataFrame Column'] = pd.to_datetime(df['DataFrame Column'], format=specify your format)
life_expectancy_df['Year'] = pd.to_datetime(life_expectancy_df['Year'], format='%Y', errors='coerce')

life_expectancy_df.info()
```

```python
life_expectancy_df.describe()
```

```python
# review distribution of each "feature" using .hist method
life_expectancy_df.hist(bins=50, figsize=(25,20))
plt.show()
```

```python
# examine correlations 
corr_matrix = life_expectancy_df.corr()

# examine how each attribute correlates with life expectancy
# note corr only meaures linear correclations
corr_matrix["Life expectancy"].sort_values(ascending=False) # sort in descending order
```

```python
# create a short list of attributes to examine using scatter_matrix method
attributes = [
    'Life expectancy',
    'Schooling',
    'Income composition of resources',
    'BMI',
    'GDP',
    'Alcohol',
    'percentage expenditure',
    'Population', 
    'infant deaths'
]
```

```python
# examine scatter plot of numberical attribute agains every other  numerical attibute
scatter_matrix(life_expectancy_df[attributes], figsize=(25,20), diagonal='kde')
plt.show()
```

```python
# look again for missing values
life_expectancy_df.info()
```

```python
# clean data by dropping nulls
life_expectancy_df.dropna(subset=["Schooling"], inplace=True)
```

```python
life_expectancy_df.info()
```

```python
# group by country
life_expectancy_df_grouped = life_expectancy_df.groupby(['Country', 'Year']).mean()
life_expectancy_df_grouped
```
